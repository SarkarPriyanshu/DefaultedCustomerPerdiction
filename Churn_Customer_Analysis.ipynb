```python
FinalFeaturesToDrop = list()

def showNullPercentage(data=None):
  '''
     Info : This function display the null percent of feature in a dataset

     Input: dataframe

     Output: Countplot Graph
  '''
  # Calculating the percent null values in each column
  data = round(data.isna().sum()/data.shape[0] * 100,1)\
              .reset_index()\
              .rename(columns={'index':'Features',0:'Null Percent %'})\
              .sort_values('Null Percent %',ascending=False)[:18]

  plt.figure(figsize=(20,5))

  # plot count plot
  bar_plot = sns.barplot(x='Features', y='Null Percent %',
                        data=data, palette='viridis')

  # Add title and labels
  plt.title('Percentage of Null Values in Features')
  plt.xlabel('Features')
  plt.ylabel('Null Value Percentage (%)')
  plt.ylim((0,100))
  plt.xticks(rotation=90)

  # Add percentage labels on top of each bar
  for p in bar_plot.patches:
      bar_plot.annotate(f'{p.get_height()}%',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom', fontsize=10)

  # Show the plot
  plt.show()
```

## Feature Selections

### Chi Square Test

- Applying Chi square test to check association between Target and Categorical Features
- H0: Cat Feature and Target are Independent of each other. (No association)
- H0: Cat Feature and Target are Dependent on each other. (Association)
- Alpha (Significance Level): 0.05

```python
for cat in categorical_features:
  if cat != 'Approved_Flag':
    statistic, pvalue , _, _ =chi2_contingency(pd.crosstab(df[cat],df['Approved_Flag']))
    print(f'FeatureName: {cat}')
    print(f'Statistic: {statistic}, P-value: {pvalue}')
    print(f'Result: {"Associated (Enough proof to reject H0)" if pvalue<0.05 else "Not Associated (Failed to reject H0)"}\n')
```

### Anova Test
- ttest to check if the mean of 2 class (p & e) are equal or not equal for all numerical features.
-  H0: the means of the samples are equal.
-  H1: the means of the samples are unequal.
- alpha : 0.05 confidence-l : 0.95%

```python
FeaturesToDrop = list()
```

```python
groups = list()
for feature in df.select_dtypes(exclude='object').columns.tolist():
  group = list()

  for category in df['Approved_Flag'].unique():
    group = [value for value,cat in zip(df[feature],df['Approved_Flag']) if cat == category]
    groups.append(group)
  f_statistic, p_value = f_oneway(*groups)
  print(f"Feature : {feature:{20}} Result : {f'Reject H0 (Not equal, p-value:{p_value})' if p_value < 0.05 else f'Accept H0 (Not equal, p-value:{p_value})'}" )
  if p_value > 0.05:
    FeaturesToDrop.append(feature)
  groups = list()
```

### Variance Inflation Fator
- To remove multi-colinearity between Independent numerical features


```python
numeric_columns = df.select_dtypes(exclude='object').columns.to_list()
vif_data = df[numeric_columns].dropna(axis=0)
total_columns = vif_data.shape[1]
column_index = 0
FeaturesToDrop = list()

for i in range (0,total_columns):

    vif_value = variance_inflation_factor(vif_data, column_index)
    print (numeric_columns[i],'---',vif_value)


    if vif_value <= 5:
        column_index = column_index+1

    else:
        FeaturesToDrop.append(numeric_columns[i])
        vif_data = vif_data.drop([ numeric_columns[i] ] , axis=1)

del vif_data
FinalFeaturesToDrop.extend(FeaturesToDrop)
del FeaturesToDrop
```

```python
rs = RobustScaler()

X_train_tm[continuesFeatures] =  rs.fit_transform(X_train_tm[continuesFeatures])
X_val_tm[continuesFeatures] =  rs.transform(X_val_tm[continuesFeatures])
```

```python
model = Sequential()

model.add(Dense(128, input_dim=X_train_tm.shape[1],kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu',kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu',kernel_initializer='he_normal'))
model.add(BatchNormalization())
model.add(Dense(4, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
history = model.fit(X_train_tm, y_train, validation_data=(X_val_tm, y_val), epochs=50)
```

```python
plt.figure(figsize=(20,5))
plt.plot(history.history['accuracy'],label='Traning accuracy')
plt.plot(history.history['val_accuracy'],label='Validation accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(history.history['loss'],label='Traning loss')
plt.plot(history.history['val_loss'],label='Validation loss')
plt.legend()
plt.show()
```

```python
X_test_tm = pipe.transform(X_test)
X_test_tm[continuesFeatures] =  rs.transform(X_test_tm[continuesFeatures])
```

```python
y_prob = model.predict(X_test_tm)
y_pred = y_prob.argmax(axis=1)
print(classification_report(y_test,y_pred))
```
